```python
df_list = []  # empty dataframe to collect state data 

for state in us.states.STATES:
    if state.abbr in ['AK', 'HI']:
        print(f'no kinsa health data for state {state.abbr}')
    
    else:
        print(f'https://static.kinsahealth.com/{state.abbr}_data.json')
        
        r = requests.get(
            f'https://static.kinsahealth.com/{state.abbr}_data.json'
        )

        data = r.json()

        df = pd.DataFrame(data['data'], columns=data['columns'])

        df_list.append(df)  # add state df to df_list

        time.sleep(1)  # be nice :) 
    
df = pd.concat(df_list, sort=True)
```

```python
# reorder columns

df = df[[
    'region_id',
    'region_name',
    'date', 
    'anomaly_diff',
    'atypical_ili',
    'forecast_expected',
    'forecast_lower',
    'forecast_upper', 
    'observed_ili',    
]]
```

```python
df['anomaly_diff'] = df['anomaly_diff'].fillna(df['anomaly_diff'].mode()[0])
```

```python
df['atypical_ili'] = df['atypical_ili'].fillna(df['atypical_ili'].mode()[0])
```

```python
df['forecast_expected'] = df['forecast_expected'].fillna(df['forecast_expected'].mode()[0])
df['forecast_lower'] = df['forecast_lower'].fillna(df['forecast_lower'].mode()[0])
df['forecast_upper'] = df['forecast_upper'].fillna(df['forecast_upper'].mode()[0])
df['observed_ili'] = df['observed_ili'].fillna(df['observed_ili'].mode()[0])
```

```python
# Converting categorical variables to unique integers in the dataset df

anomaly_locs = sorted(df['anomaly_diff'].unique())
anomaly_locs_mapping = dict(zip(anomaly_locs, range(0, len(anomaly_locs) + 1)))
anomaly_locs_mapping
df['anomaly_diff'] = df['anomaly_diff'].map(anomaly_locs_mapping).astype(int)


atypical_locs = sorted(df['atypical_ili'].unique())
atypical_locs_mapping = dict(zip(atypical_locs, range(0, len(atypical_locs) + 1)))
atypical_locs_mapping
df['atypical_ili'] = df['atypical_ili'].map(atypical_locs_mapping).astype(int)

expected_locs = sorted(df['forecast_expected'].unique())
expected_locs_mapping = dict(zip(expected_locs, range(0, len(expected_locs) + 1)))
expected_locs_mapping
df['forecast_expected'] = df['forecast_expected'].map(expected_locs_mapping).astype(int)

lower_locs = sorted(df['forecast_lower'].unique())
lower_locs_mapping = dict(zip(lower_locs, range(0, len(lower_locs) + 1)))
lower_locs_mapping
df['forecast_lower'] = df['forecast_lower'].map(lower_locs_mapping).astype(int)


upper_locs = sorted(df['forecast_upper'].unique())
upper_locs_mapping = dict(zip(upper_locs, range(0, len(upper_locs) + 1)))
upper_locs_mapping
df['forecast_upper'] = df['forecast_upper'].map(upper_locs_mapping).astype(int)

observed_locs = sorted(df['observed_ili'].unique())
observed_locs_mapping = dict(zip(observed_locs, range(0, len(observed_locs) + 1)))
observed_locs_mapping
df['observed_ili'] = df['observed_ili'].map(observed_locs_mapping).astype(int)
```

```python
X = df.drop(['region_name'], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
# Standardizing the features
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents, 
                           columns = ['principal component 1', 'principal component 2'])
```

```python
# Separating out the target
Region_name = df.loc[:,['region_name']].values
```

```python
principalDf['Region_name'] = Region_name
```

```python
principalDf.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Cullman County', 'Chilton County', 'Jackson County']
colors = ['r', 'g', 'b']
for Region_name, color in zip(targets,colors):
    indicesToKeep = principalDf['Region_name'] == Region_name
    ax.scatter(principalDf.loc[indicesToKeep, 'principal component 1']
               , principalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 10)
ax.legend(targets)
ax.grid()
```
